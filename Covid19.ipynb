# Data Cleaning
As we an see there are many useless columns like Latitude and Longitude which we don't need for our analysis 

So, We will clean them for better results

```
df = corona_dataset.drop(["Lat","Long"],axis=1) #It will delete both the Latitude and Longitude columns
```

```
df.head(10)
```

# Aggregating the rows by country for better analysis

```
df_agg = df.groupby("Country/Region").sum()   
df_agg.head(10)
```

```
df_agg.shape #see how he reduced the shape of our dataframe
```

# Data Visualization and Comparison of some countries

```
df_agg.loc['India'].plot()
plt.show()
```

```
df_agg.loc['India'].plot()
df_agg.loc['China'].plot()
df_agg.loc['Spain'].plot()
plt.legend()
plt.show()
```

```
df_agg.loc['India'].diff().plot() # To find derivative 
plt.show()
```
